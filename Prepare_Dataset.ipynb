```python
import csv
import json

fisier_intrare = "surzhyk_toxicity.csv"
fisier_csv_iesire = "doar_propozitii.csv"
fisier_json_iesire = "import_adnotare.json"

date_json = []

print("Citesc datele si pregatesc fisierele pentru Labelbox...")

with open(fisier_intrare, "r", encoding="utf-8") as f_in, \
     open(fisier_csv_iesire, "w", encoding="utf-8", newline='') as f_csv:

    reader = csv.reader(f_in)
    writer = csv.writer(f_csv)

    # sar peste capul de tabel vechi din csv
    header = next(reader, None)

    # pun un cap de tabel nou pentru csv-ul meu curat
    writer.writerow(["Propozitie Originala"])

    for i, rand in enumerate(reader):
        # ignor randurile goale, in caz ca apar
        if not rand:
            continue

        propozitie = rand[0]

        # 1. scriu propozitia in noul csv
        writer.writerow([propozitie])

        # 2. adaug propozitia in formatul pentru json
        # ca sa mearga importul in labelbox, folosesc 'row_data' si generez eu un id
        date_json.append({
            "row_data": propozitie,
            "external_id": f"prop_{i+1}"
        })

# salvez fisierul json
with open(fisier_json_iesire, "w", encoding="utf-8") as f_json:
    json.dump(date_json, f_json, ensure_ascii=False, indent=4)

print(f"Gata! Am extras {len(date_json)} propozitii.")
print(f"Fisierele mele salvate sunt: '{fisier_csv_iesire}' si '{fisier_json_iesire}'.")
```
